```python
location_parameters = {
    67569:['um050', 'pressure', 'humidity', 'temperature', 'um005', 'um003', 'pm1', 'um025', 'um010', 'um100', 'pm25', 'pm10'], #Bengalore
    288233:["pm25","um100","pressure","um010","um025","pm10","um003","humidity","um005","pm1","temperature","um050"], # Kalol, Gandhi Nagar
    64931:["um100","um050","um005",'pm10','um025','pm1','pm25','um003','um010'], #Bhatinda
    66673:['um003','um003','pm10','um005','pm25','um010''um100''pressure','um050','um025','pm1','temperature','humidity'], #Hisar
    8118:["pm25"], # New Delhi
    62543:['pressure','pm25','um010','humidity','um003','temperature','um100','um025','um050','um005','pm10','pm1'], # Greater Kailash 2 
    1667903:['um010','humidity','temperature','um050','pm1','um003','um005','pm25','pm10','um025','pressure','um100'],  # 15 Oak Drive Outdoor
    362098:['temperature','um050','pressure','um003','um005','humidity','um025','um100','um010','pm10','pm1','pm25'], # Greater Noida
    67569:['pm10','um010','um100','um050','um005','pressure','um025','um003','temperature','pm1','humidity','pm25'], #Tarkeshwar, West Bengal
    8172:['pm25'], #Kolkata
    220704:['temperature','um010','pm10','pm25','pressure','um025','humidity','temperature','um003','um100','um050','um005','pm1'], #Kharagpur, West Bengal
    8039:['pm25'], # Mumbai
    8557:['pm25'], #Hyderabad
    63704:['um003','um025','pm10','pm1','humidity','temperature','um010','um005','um050','pm25','um100','pressure'], #Madikeri, Karnataka
    229138:['um050','pm10','temperature','humidity','pm1','um003','pressure','um025','um005','um010','um100','pm25'], # Srinivaspur, Karnataka
    8558:['pm25'] #Chennai
}

govt_sensors = set([8039,8118,8172,8557,8558])

for location_id in location_parameters.keys():
    # make the directory of location_id
    os.makedirs(f'data/2024/02/02/{location_id}')
    for parameter in location_parameters[location_id]:
        if location_id in govt_sensors:
            # govt
            url = f"https://api.openaq.org/v2/measurements?location_id={location_id}&parameter={parameter}&date_from=2024-02-02T00:00:00+05:30&date_to=2024-02-03T00:00:00+05:30"
            response = requests.get(url)
            if response.status_code == 200:
                print('govertment hit')
                # Access the response data in JSON format
                data = response.json()
                # check if you've got some data or not
                if(len(data['results'])>0):
                    # print(data['results'])
                    with open(f'data/2024/02/02/{location_id}/{parameter}.json','w') as file:
                        json.dump(data['results'],file,indent=2)
            else:
                print(f"Error in govt hit: {response.status_code}")
        else:
            # community 
            start_timestamp = datetime(2024, 2, 2, 0, 0, 0)
            num_intervals = 8 
            flag=False
            for i in range(num_intervals):
                current_timestamp1 = start_timestamp + timedelta(hours=3 * i)
                timestamp1 = current_timestamp1.strftime('%Y-%m-%dT%H:%M:%S')
                current_timestamp2 = current_timestamp1 + timedelta(hours=3)
                timestamp2 = current_timestamp2.strftime('%Y-%m-%dT%H:%M:%S')
                url = f"https://api.openaq.org/v2/measurements?location_id={location_id}&parameter={parameter}&date_from={timestamp1}+05:30&date_to={timestamp2}+05:30"
                response = requests.get(url)
                if response.status_code == 200:
                    print('community hit')
                    # Access the response data in JSON format
                    data = response.json()
                    # check if you've got some data or not
                    if(len(data['results'])>0):
                        if flag==False:
                            os.mkdir(f'data/2024/02/02/{location_id}/{parameter}')
                            flag=True
                        with open(f'data/2024/02/02/{location_id}/{parameter}/{timestamp1[10:]}_{timestamp2[10:]}.json','w') as file:
                            json.dump(data['results'], file,indent=2)
                else:
                    print(f"Error in community hit: {response.status_code}")
        time.sleep(2)
```

## Reading the downloaded data , applying transformations and saving to HDFS.

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *

# creating a spark session
spark = SparkSession.builder\
    .appName("warehouse_dump")\
    .getOrCreate()
```

```python
from pyspark.sql.utils import AnalysisException

locationIds = [1667903,229138,362098,63704,66673,8039,8172,8558,220704,288233,62543,64931,67569,8118,8557]

for locationId in locationIds:
    # get the list of parameters
    parameters = os.listdir(f'data/2024/02/02/{locationId}')
    for parameter in parameters:
        try:
            data = spark.read\
                    .option("multiline","true") \
                    .json(f'data/2024/02/02/{locationId}/{parameter}')
            final_df = data.select('locationId','date.utc','parameter','value')
            final_df.write.format("orc")\
                .mode('append').save("hdfs://localhost:9000/user/OpenAQ/data/input")
        except AnalysisException:
            print(f'inferSchema faiied for {locationId}/{parameter}')
```

### building latest table

```python
schema="locationId INT, utc_time TIMESTAMP, parameter STRING, value DOUBLE"
df = spark.read\
    .option('schema',schema)\
    .orc('hdfs://localhost:9000/user/OpenAQ/data/input')
df.show(truncate=False)
```

```python
df.printSchema()
```

### Select the data of only recent 2 years

```python
recent_df = df\
    .filter(year('local').between(year(current_date()) - 1, year(current_date())))

recent_df.show(truncate=False)
```

### Dump to HDFS

```python
recent_df.write\
    .option('header',True)\
    .format("csv").mode('overwrite')\
    .save("hdfs://localhost:9000/user/OpenAQ/data/latest")
```
