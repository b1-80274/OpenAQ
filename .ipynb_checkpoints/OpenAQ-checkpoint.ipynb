```python
from datetime import datetime, timedelta
import os

def save_monthly_measurements():
    cities=['New Delhi','Mumbai','Kolkata','Hyderabad','Chennai']

    for i in range(1,32):
            directory_path = f'/home/sad7_5407/Desktop/Data Engineering/data/india/2023/01/{i}'
            os.makedirs(directory_path)
        
            start_date = datetime(2023, 1, i)
            end_date = start_date + timedelta(days=1)
            start_date = start_date.strftime("%Y-%m-%d")
            end_date = end_date.strftime("%Y-%m-%d")
            for city in cities:
                    
                    file=open(f'data/india/2023/01/{i}/{city}.json','a')                
                    def api_hit():
                        status, resp = api.measurements(country='IN',location=city,date_from=start_date,date_to=end_date)
                        print(status)
                        # Write the list of dictionaries to the JSON file
                        json.dump(resp['results'], file, indent=2)
                        time.sleep(2)
                        
                        
                    try:
                        api_hit()
                    except requests.exceptions.Timeout:
                        print("APIerror 408 occured.\n connection terminated .\n sleeping for 10 seconds")
                        time.sleep(10)
                        api_hit()
                    except Exception as e:
                        print(f"An error occurred: {e}")
                        time.sleep(5)

                        
                    
                    
save_monthly_measurements()
```

```python
schema="city STRING, local_time TIMESTAMP, parameter STRING, unit STRING, value DOUBLE"
df = spark.read\
    .csv('hdfs://localhost:9000/user/OpenAQ/data/input',schema=schema)
df.show(truncate=False)
```

### Select the data of only recent 2 years

```python
recent_df = df\
    .filter(year('local_time').between(year(current_date()) - 1, year(current_date())))

recent_df.show(truncate=False)
```

### Dump to HDFS

```python
recent_df.write\
    .format("csv").mode('overwrite')\
    .save("hdfs://localhost:9000/user/OpenAQ/data/latest")
```
